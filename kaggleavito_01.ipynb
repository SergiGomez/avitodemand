We have seen that every observation is a different item_id, but one user may sell several items

For our first prediction, we will only keep variables easy to be processed by a simple Machine Learning model, such as Random Forest or even a Logistic Regression. We will call this first dataset 'train_1'

```python
train_1 = train[['item_id','user_id','region','city','parent_category_name','category_name','param_1',
                 'param_2','param_3','price','item_seq_number','activation_date','user_type','deal_probability']]
print("Head of our new dataset train_1 :",train_1.head(n = 1))
```

I will tansform the item_id and user_id to make it easier to read and process so we can save up some memory

```python
# Let me take only the price and user_id to fit a very simple Logistic Regression
X = train[['price']]
y = train[['deal_probability']]
```

```python
X_imput = Imputer(missing_values='NaN').fit_transform(X, y)
```

```python
nanPrice_tuple_imput = np.where(np.isnan(X_imput))
print("Percentage (%) of NaNs in the price after imputation: ", len(nanPrice_tuple_imput[0])*100 / X_imput.shape[0])
```

```python
# Split in train and test set
X_train, X_test, y_train, y_test = train_test_split(X_imput, y, test_size=0.20, random_state=42)
print("Shape X_train: ", X_train.shape)
print("Shape y_train: ", y_train.shape)
print("Shape X_test: ", X_test.shape)
print("Shape y_test: ", y_test.shape)
```
